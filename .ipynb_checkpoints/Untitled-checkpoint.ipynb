```python
import  requests
import json
from pprint import pprint
import pandas as pd
```

# Data Recorded in Google Dictionary Extension

```python
with open('../GoogleDictionaryHistory.csv', 'r') as f:
    data = {}
    for line in f.readlines():
        line = line.split('\t')
        data[line[2]] = line[-1].strip()
```

# Oxford Dictionary API

```python
app_id = '07cd29e8'
app_key = '85464a0c2501389da79937f504e5a05d'
```

```python
language = 'en'
word = 'vehement'
url = 'https://od-api.oxforddictionaries.com/api/v1'
sentences = '/entries/{}/{}/sentences'
```

```python
res = []
for word in data:
    r = requests.get(url + sentences.format(language, word), headers = {'app_id' : app_id, 'app_key' : app_key})
    if r.status_code == 404: continue
    t = r.json()['results'][0]['lexicalEntries'][0]['sentences']
    d = {
        'word': word,
        'example': [s['text'] for s in list(filter(lambda x: x['regions'] == ['North American'], t))]
    }
    res.append(d)
```

# Google Dictionary API

```python
base_url = 'https://googledictionaryapi.eu-gb.mybluemix.net/?define='
```

```python
examples = {}
for word in data:
    r = requests.get(base_url + word)
    examples[word] = r.json()
```

```python
res = []
for word in examples:
    for key in examples[word]['meaning']:
        for item in examples[word]['meaning'][key]:
            elm = {
                'word': word,
                'definition': item['definition'],
                'pos': key
            }
            if 'example' in item:
                elm['example'] = item['example']
            if 'synonyms' in item:
                elm['synonyms'] = item['synonyms']
            res.append(elm)
```

```python
df = pd.DataFrame(res)
```
